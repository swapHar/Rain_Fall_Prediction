```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Read the data from data file

```python
data = pd.read_csv('./data/austin_weather.csv')
data.head()
```

Learn about our data

```python
data.dtypes
```

```python
data.shape
```

```python
# Drop few columns
data = data.drop(['Events', 'SeaLevelPressureHighInches', 'SeaLevelPressureLowInches'], axis=1)
```

```python
print(data['DewPointHighF'].unique())
```

```python
print(data['PrecipitationSumInches'].unique())
```

As we know from dataset information coloumn 'PrecipitationSumInches' contains 'T' as value at some instances.
we need to convert all object datatypes to int or float datatype as we need all values in numerical format.
Also values in Date column to date format.

```python
data = data.replace('T','')
data = data.replace('-','')
```

```python
data['DewPointHighF'] = pd.to_numeric(data['DewPointHighF'])
data['DewPointAvgF'] = pd.to_numeric(data['DewPointAvgF'])
data['DewPointLowF'] = pd.to_numeric(data['DewPointLowF'])
data['HumidityHighPercent'] = pd.to_numeric(data['HumidityHighPercent'])
data['HumidityAvgPercent'] = pd.to_numeric(data['HumidityAvgPercent'])
data['HumidityLowPercent'] = pd.to_numeric(data['HumidityLowPercent'])
data['SeaLevelPressureAvgInches'] = pd.to_numeric(data['SeaLevelPressureAvgInches'])
data['VisibilityHighMiles'] = pd.to_numeric(data['VisibilityHighMiles'])
data['VisibilityAvgMiles'] = pd.to_numeric(data['VisibilityAvgMiles'])
data['VisibilityLowMiles'] = pd.to_numeric(data['VisibilityLowMiles'])
data['WindHighMPH'] = pd.to_numeric(data['WindHighMPH'])
data['WindAvgMPH'] = pd.to_numeric(data['WindAvgMPH'])
data['WindGustMPH'] = pd.to_numeric(data['WindGustMPH'])
data['PrecipitationSumInches'] = pd.to_numeric(data['PrecipitationSumInches'])
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.isnull().sum()
```

Remove rows from data with missing values

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.describe()
```

Lets split date in year and month for further analysis

```python
data['Year'] = pd.DatetimeIndex(data['Date']).year
data['Month'] = pd.DatetimeIndex(data['Date']).month
```

```python
data = data.drop('Date', axis=1)
```

```python
print(data.columns)
```

```python
x = data.drop(columns=['PrecipitationSumInches','Year'])
y = data['PrecipitationSumInches']
```

```python
#standardiz the features in DataFrame x using StandardScaler, 
# and then converts the standardized features back to a DataFrame.
sc = StandardScaler()
x_standard = sc.fit_transform(x)
x_standard = pd.DataFrame(x_standard,columns=x.columns)
```

```python
# Split the dataset
x_train,x_test,y_train,y_test = train_test_split(x_standard, y, test_size=0.2, random_state=42)
```

Create the linear regression model and train it

```python
rain_model = LinearRegression()
rain_model.fit(x_train, y_train)
```

```python
y_predicted = rain_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print('Mean absolute error = ', mean_absolute_error(y_test, y_predicted))
print('Mean squared error = ', mean_squared_error(y_test, y_predicted))
print('R2 score = ', r2_score(y_test, y_predicted))
```

```python
from sklearn.metrics import accuracy_score

y_test = np.array(y_test).astype(int)
y_predicted = np.array(y_predicted).astype(int)
accuracy = accuracy_score(y_test, y_predicted)

print("Accuracy:", accuracy)
```
